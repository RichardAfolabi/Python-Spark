```python
# Read the CSV file
df = sqlContext.read.format("com.databricks.spark.csv").options(header='true', inferschema="true").load("csv/viewership_data.csv")
df.count()
```

> Register a `TempTable` to make `Spark SQL` work like SQL - display structure and schema

```python
df.registerTempTable("raw_data")
df.printSchema()
```

```python
sqlContext.sql(
    "SELECT dt, evar31, rptg_st_name, post_evar32, connection_type, post_evar11 \
    FROM raw_data \
    WHERE rptg_st_name LIKE '%DTVAPP%' \
    AND post_evar32 <> '' \
    LIMIT 10").show()
```

```python
sqlCtx.sql(
    "SELECT DISTINCT post_evar31, count(*) AS COUNT \
    FROM raw_data GROUP BY post_evar31 ORDER BY COUNT DESC").show()
```
